```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=0,
    separator = "."
)

chunks = splitter.split_text(text)
```

```python
len(chunks)
```

```python
for chunk in chunks:
    print(len(chunk))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
r_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " "],
    chunk_size = 200,
    chunk_overlap = 0
)

chunks = r_splitter.split_text(text)
```

### Converting sentence to vectors using sentence transformer

```python
from sentence_transformers import SentenceTransformer
```

```python
df.text
```

```python
encoder = SentenceTransformer(
    "all-mpnet-base-v2"
)
vectors = encoder.encode(df.text)
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
import faiss
```

```python
vectors[1].shape[0]
```

```python
dim = vectors[1].shape[0]
```

```python
dim
```

```python
index = faiss.IndexFlatL2(dim)
```

```python
index
```

```python
index.add(vectors)
```

```python
search_query = "I feel energetic"
```

```python
vec = encoder.encode(search_query)
```

```python
vec.shape
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
```

```python
svec.shape
```

```python
index.search(svec, k=2)
```

```python
distance, I = index.search(svec, k=2)
```
